## La pile d´appel de fonctions

Testez l'execution des fonctions A et B:

```python
def A():
    for i in range (5):
        print('A'+str(i))
    print('-> fin de la fonction A')
A()
```

```python
def B():
    for i in range (5):
        print('B'+str(i))
        if i==2:
            print('-> appel de la fonction A')
            A()
    print('-> fin de la fonction B')
B()
```

Commentons le résultat obtenu : la `fonction A` est invoquée au milieu de l'exécution de la `fonction B`. Cela semble naturel mais pour la machine, ce mécanisme est loin d'être simple : `B` interrompt son exécution pour exécuter `A`. Lorsque `A` se termine, on remarque que `B` reprend son exécution là ou elle l'avait interrompu. 

La variable `i` n'a pas été perturbée par l'exécution de `A` et qui continue à être incrémentée comme si rien ne s'était passé.

Ce comportement est logique et naturel mais nécessite la mise en œuvre d'un mécanisme complexe dans la machine : **la pile d'exécution**.

Observons l´exécution de B avec [Python tutor](https://pythontutor.com/visualize.html#code=def%20A%28%29%3A%0A%20%20%20%20for%20i%20in%20range%20%285%29%3A%0A%20%20%20%20%20%20%20%20print%28'A'%2Bstr%28i%29%29%0A%20%20%20%20print%28'-%3E%20fin%20de%20la%20fonction%20A'%29%0A%0Adef%20B%28%29%3A%0A%20%20%20%20for%20i%20in%20range%20%285%29%3A%0A%20%20%20%20%20%20%20%20print%28'B'%2Bstr%28i%29%29%0A%20%20%20%20%20%20%20%20if%20i%3D%3D2%3A%0A%20%20%20%20%20%20%20%20%20%20%20%20print%28'-%3E%20appel%20de%20la%20fonction%20A'%29%0A%20%20%20%20%20%20%20%20%20%20%20%20A%28%29%0A%20%20%20%20print%28'-%3E%20fin%20de%20la%20fonction%20B'%29%0A%0AB%28%29&cumulative=false&heapPrimitives=nevernest&mode=edit&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false).

Lorsqu'une fonction est exécutée, les informations nécessaire à son fonctionnement (ce qu'on appelle son contexte d'exécution) sont sauvegardées dans une pile d'exécution. Celle pile va donc en particulier contenir la **variable locale i**.

L'exécution de la `fonction B` va se dérouler en 3 phases :

1. Début de la `fonction B`, on empile un nouveau contexte pour la `fonction B`.
<img src='http://www.fredpeuriere.com/NSI-term/PROGRAMMATION/pile1.png' width=100/>
2. La `fonction A` est invoquée. Un nouveau contexte est empilé sur la pile.
<img src='http://www.fredpeuriere.com/NSI-term/PROGRAMMATION/pile2.png' width=100/>
3. La `fonction A` se termine. Son contexte est alors dépilé et `B` retrouve son contexte et peut ainsi reprendre là où il en était.  
<img src='http://www.fredpeuriere.com/NSI-term/PROGRAMMATION/pile3.png' width=100/>  

Notons que le programme ne se termine que lorsque la **pile d´exécution est vide**.

<span style="color:white;background:green;font-size: 18px;font-weight: bold;">Un premier bilan:</span>

On parle de **récursivité** lorsqu´une fonction s´appelle elle-même. 

Il faut toujours faire attention à ce que les **appels récursifs** se terminent. Dans cet exemple, il est assez simple de s'en persuader : à chaque appel récursif, l'argument `n` est décrémenté d'une unité. Cette valeur finira donc par être égale à 0, et dans ce cas les appels s'arrêtent et les différents contextes d´exécution des fonction sont dépilés en reprenant là où ils s´étaient arrêtés.

La condition d´arrêt est appelé **cas de base**, elle est nécessaire pour que les appels récursifs ne soient pas infinis comme dans l´**effet Droste**. <br> 
Pour éviter cette situation, chaque langage de programmation impose une limite à la capacité de la pile d´appel. Si elle est dépassée (stack *overflow*), le programme est stoppé et une erreur est retournée.

```python
# Avec python, on accède ainsi à la capacité de la pile:
import sys
print(sys.getrecursionlimit())
# Cette valeur est modifiable
```

Un algorithme est dit **récursif** lorsqu´il utilise une fonction qui s´appelle elle même.

Cette technique de programmation est utile pour résoudre des problèmes complexes qui nécessitent un retour arrière:

- Colorier une carte de façon à ce que deux pays voisins n’aient pas la
même couleur
- Résoudre une grille de Sudoku
- Dessiner des fractales

... ou parcourir des structures de données complexes:
- Les arbres
- Les graphes

Pour écrire un algorithme récursif on procède en deux temps:
- on définit un **cas de base**
- on établit une relation de **récurrence**
